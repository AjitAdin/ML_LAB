```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
from  sklearn.datasets import fetch_california_housing
```

```python
california_data=fetch_california_housing(as_frame=True)
# Convert to DataFrame
data = california_data.frame
# Display the first few rows of the DataFrame
data.head()
```

```python
corelation_matrix=data.corr()
corelation_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corelation_matrix,annot=True,fmt='.2f',cmap='coolwarm')
plt.title('Correlation Matrix of California Housing Dataset')
plt.show()
```

```python
sns.pairplot(data,hue='MedHouseVal', palette='coolwarm')
plt.suptitle('Pairplot of California Housing Dataset', y=1.02)
```
